Setup

```python
import yaml

# Define the content for the YAML file
data = {
    'train': 'wb_localization_dataset/images/train',
    'val': 'wb_localization_dataset/images/val',
    'nc': 1,
    'names': {0: 'word'}
}

# Define the path to the output YAML file
output_file_path = 'custom_data.yaml'

# Write the content to the YAML file
with open(output_file_path, 'w') as file:
    yaml.dump(data, file, default_flow_style=False)
```

```python

!yolo train model=yolov8x.pt data=/content/custom_data.yaml epochs=100 batch=2 imgsz=640
```
